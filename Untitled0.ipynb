```python
MAX_ALUNOS = 20


class Student:
    def __init__(self, nome, disciplina, matricula, senha):
        self.nome = nome
        self.disciplina = disciplina
        self.matricula = matricula
        self.senha = senha

    def __str__(self):
        return (
            f"Nome: {self.nome} | "
            f"Disciplina: {self.disciplina} | "
            f"Matrícula: {self.matricula}"
        )


class HashTable:
    def __init__(self, tamanho=31, metodo_hash=1):
        self.tamanho = tamanho
        self.tabela = [[] for _ in range(tamanho)]
        self.metodo_hash = metodo_hash

    def hash1(self, chave):
        return sum(ord(c) for c in chave) % self.tamanho

    def hash2(self, chave):
        valor = 0
        for c in chave:
            valor = (valor * 31 + ord(c)) % self.tamanho
        return valor

    def calcular_hash(self, chave):
        if self.metodo_hash == 1:
            return self.hash1(chave)
        return self.hash2(chave)

    def inserir(self, senha, indice_aluno):
        pos = self.calcular_hash(senha)

        for i, (s, _) in enumerate(self.tabela[pos]):
            if s == senha:
                self.tabela[pos][i] = (senha, indice_aluno)
                return

        self.tabela[pos].append((senha, indice_aluno))

    def buscar(self, senha):
        pos = self.calcular_hash(senha)

        for s, indice in self.tabela[pos]:
            if s == senha:
                return indice
        return None

    def remover(self, senha):
        pos = self.calcular_hash(senha)

        for i, (s, _) in enumerate(self.tabela[pos]):
            if s == senha:
                del self.tabela[pos][i]
                return True
        return False


class StudentManager:
    def __init__(self):
        self.alunos = []
        self.hash_table = HashTable()

    def ordenar_por_matricula(self):
        self.alunos.sort(key=lambda aluno: aluno.matricula)

    def atualizar_indices_hash(self):
        self.hash_table = HashTable()

        for i, aluno in enumerate(self.alunos):
            self.hash_table.inserir(aluno.senha, i)

    def criar_aluno(self, nome, disciplina, matricula, senha):
        if len(self.alunos) >= MAX_ALUNOS:
            print("Limite de alunos atingido.")
            return

        aluno = Student(nome, disciplina, matricula, senha)
        self.alunos.append(aluno)
        self.ordenar_por_matricula()
        self.atualizar_indices_hash()
        print("Aluno cadastrado com sucesso!")

    def listar_alunos(self):
        if not self.alunos:
            print("Nenhum aluno cadastrado.")
            return

        for aluno in self.alunos:
            print(aluno)

    def busca_binaria(self, matricula):
        self.ordenar_por_matricula()

        esquerda = 0
        direita = len(self.alunos) - 1

        while esquerda <= direita:
            meio = (esquerda + direita) // 2

            if self.alunos[meio].matricula == matricula:
                return meio

            if self.alunos[meio].matricula < matricula:
                esquerda = meio + 1
            else:
                direita = meio - 1

        return -1

    def buscar_aluno(self, matricula):
        indice = self.busca_binaria(matricula)

        if indice != -1:
            return self.alunos[indice]

        return None

    def atualizar_aluno(self, matricula):
        aluno = self.buscar_aluno(matricula)

        if not aluno:
            print("Aluno não encontrado.")
            return

        novo_nome = input("Novo nome: ")
        nova_disciplina = input("Nova disciplina: ")

        aluno.nome = novo_nome
        aluno.disciplina = nova_disciplina

        print("Aluno atualizado com sucesso!")

    def excluir_aluno(self, matricula):
        indice = self.busca_binaria(matricula)

        if indice == -1:
            print("Aluno não encontrado.")
            return

        senha = self.alunos[indice].senha

        del self.alunos[indice]
        self.hash_table.remover(senha)
        self.atualizar_indices_hash()

        print("Aluno removido com sucesso!")

    def login(self, senha):
        indice = self.hash_table.buscar(senha)

        if indice is not None:
            print("Login realizado com sucesso!")
            print(self.alunos[indice])
        else:
            print("Senha inválida.")


def menu():
    sistema = StudentManager()

    while True:
        print("\n=== GERENCIAMENTO DE ALUNOS ===")
        print("1 - Cadastrar aluno")
        print("2 - Listar alunos")
        print("3 - Buscar aluno por matrícula")
        print("4 - Atualizar aluno")
        print("5 - Excluir aluno")
        print("6 - Login por senha")
        print("0 - Sair")

        opcao = input("Escolha: ")

        if opcao == "1":
            nome = input("Nome: ")
            disciplina = input("Disciplina: ")
            matricula = int(input("Matrícula: "))
            senha = input("Senha: ")

            sistema.criar_aluno(nome, disciplina, matricula, senha)

        elif opcao == "2":
            sistema.listar_alunos()

        elif opcao == "3":
            matricula = int(input("Matrícula: "))
            aluno = sistema.buscar_aluno(matricula)

            if aluno:
                print(aluno)
            else:
                print("Aluno não encontrado.")

        elif opcao == "4":
            matricula = int(input("Matrícula: "))
            sistema.atualizar_aluno(matricula)

        elif opcao == "5":
            matricula = int(input("Matrícula: "))
            sistema.excluir_aluno(matricula)

        elif opcao == "6":
            senha = input("Senha: ")
            sistema.login(senha)

        elif opcao == "0":
            print("Encerrando sistema.")
            break

        else:
            print("Opção inválida.")


if __name__ == "__main__":
    menu()
```
